```python
'''
Ishank Jain
ID: 20757588
libraries
'''
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns 
import scipy as sp
from sklearn import preprocessing 
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('dataA.csv')
df.head(5)
feature = list(df)
```

```python
df = df.set_index('Unnamed: 0')
df.head(5)
```

```python
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
imp_mean.fit(df)
```

```python
arr = imp_mean.transform(df)
arr
```

```python
arr = pd.DataFrame(arr)
arr.head(5)
```

```python
arr.columns
```

```python
arr.columns =  feature[1:]
```

```python
arr.head(5)
```

```python
bins = np.linspace(-10, 10, 40)
```

```python
plt.hist(arr['fea.9'],bins, alpha=0.5, edgecolor='black',label='f 9')
plt.hist(arr['fea.24'],bins, alpha=0.5, edgecolor='black',label='f 24')
plt.legend(loc='upper right')
plt.show()
```

```python
scaler = preprocessing.MinMaxScaler()
scaler.fit(arr)
ar = scaler.transform(arr)
pd.DataFrame(ar).head(5)
```

```python
bins = np.linspace(-0.25, 1, 40)
```

```python
plt.hist(ar[9],bins, alpha=0.5, edgecolor='black',label='f 9')
plt.hist(ar[24],bins, alpha=0.5, edgecolor='black',label='f 24')
plt.legend(loc='upper right')
plt.show()
```

```python
ar=sp.stats.zscore(arr)
pd.DataFrame(ar).head()
```

```python
plt.hist(ar[9],bins, alpha=0.5, edgecolor='black',label='f_9_z-score')
plt.hist(ar[24],bins, alpha=0.5, edgecolor='black',label='f_24_z-score')
plt.legend(loc='upper right')
plt.show()
```
